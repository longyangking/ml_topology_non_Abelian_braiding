```python
import sys 
sys.path.append("..") 

import matplotlib.pyplot as plt

import numpy as np
from braiding_model import NonHermitianHamiltonianBraid, BraidOperator, NonHermitianHamiltonianTorusKnot
from braiding_model import obtain_model_braid
```

# $\sigma_1 \sigma_2 $

```python
tau1 = BraidOperator(1)
tau2 = BraidOperator(2, is_inverse=True)
braid_operators = [tau2, tau1]
n_band = 3
model = obtain_model_braid(n_band=n_band, braid_operators=braid_operators)
```

```python
ks = np.linspace(0, 2 * np.pi, 100)
Es = np.array([model.get_energies(k) for k in ks])

ax = plt.figure().add_subplot(projection='3d')
for index in range(n_band):
    x = np.real(Es[:, index])
    y = np.imag(Es[:, index])
    ax.plot(x, y, ks, label='{0}'.format(index))
    ax.legend()

plt.show()
```

```python
directory_name = './figs/Generate Braids/'
braid_name = 'sigma1_sigma2.txt'
model.save_band('{0}{1}'.format(directory_name, braid_name))
```
