```python
import sys 
sys.path.append("..") 

import matplotlib.pyplot as plt

import numpy as np
from braiding_model import NonHermitianHamiltonianBraid, BraidOperator, HamiltonianEffective, show_braid_words
```

```python
n_band = 2
braid_operators = [BraidOperator(1)]
hamiltonian_effective = HamiltonianEffective(n_band=n_band, braid_operators=braid_operators)
show_braid_words(braid_operators)

ax = plt.figure().add_subplot(projection='3d')
ks = np.linspace(0, 2 * np.pi, 10)
Es = np.array([hamiltonian_effective.get_energies(k) for k in ks])
for index in range(n_band):
    x = np.real(Es[:, index])
    y = np.imag(Es[:, index])
    ax.scatter(x, y, ks/(np.pi), label='{0}'.format(index))
    ax.legend()

ks = np.linspace(0, 2 * np.pi, 100)
Es = np.array([hamiltonian_effective.get_energies_Fourier(k) for k in ks])
for index in range(n_band):
    x = np.real(Es[:, index])#hamiltonian_effective.save_band_coefficients('2D', '1_-2')
    y = np.imag(Es[:, index])
    ax.scatter(x, y, ks/(np.pi), label='{0}'.format(index))
    ax.legend()

plt.show()

hamiltonian_effective.save_band_coefficients('figs/effective_lattice', '1')
```
